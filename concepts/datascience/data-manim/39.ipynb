```python
import pandas as pd

df = pd.read_csv("./pandas/chipo.csv")

df.head(5)
```

```python
df.loc[df["choice_description"].isnull(), "choice_description"] = "NoData"
df.head(5)
```

```python
# 방법 1
def check_data(x, val):
    return val in str(x)

df = df.loc[df["choice_description"].apply(lambda x: check_data(x, "Black"))]
df.head(5)
```

```python
# 방법 2
df = df.loc[df["choice_description"].str.contains("Black")]
df.head(5)
```
