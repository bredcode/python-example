```python
import pandas as pd

df = pd.read_csv("./pandas/chipo.csv")

df.head(5)
```

```python
df["new_price"] = df["item_price"].copy()
df["new_price"] = df["new_price"].str.replace("$","")
df["new_price"] = df["new_price"].astype("float")
df.head(5)
```

```python
df.sort_values(["new_price"], ascending=True).reset_index(drop=True)
```
